```python
import numpy as np
from bisect import insort
R = 15 # max simulator runs
time = 0
busy = False
generated_arrivals = 0
EventList = []
waiting_queue = []
calendar = []
```

```python
class event:
    def __init__(self, id, event_time, event_type, arrival_time, service_time, interarrival_time):
        self.id = id
        self.event_time = event_time
        self.event_type = event_type
        self.arrival_time = arrival_time
        self.service_time = service_time
        self.interarrival_time = interarrival_time
    def __lt__(self, other):
        return self.event_time < other.event_time
    def __repr__(self):
        return (f"ClientId: {self.id}, EventTime: {self.event_time}, EventType: {self.event_type}, ServiceTime: {self.service_time}, ArrivalTime: {self.arrival_time}, Interarrival: {self.interarrival_time}\n")


class Server:
    ARRIVAL = "arrival"
    DEPARTURE = "departure"

    def run(self):
        self._instance()
        
        while (True):
                print("BUSY: ", self._busy)
                print ("time: ", self._time)
                if(self._maximum_events>0 and self._time == 0):
                    self._total_events +=1
                    print("Primo")
                    self._genArrival()
            
            
                # for i in range(len(EventList)):
                    print(EventList)


                if(EventList):
                    e = EventList.pop(0)
                    print("Total: ",self._total_events)
                    self._total_events +=1
                    
                    #print("id:", e.id, "Time:", e.event_time, "Type: ", e.event_type)
                    self._EventHandler(e)


                if(e.event_type == self.ARRIVAL and self._generated_arrivals < self._maximum_events):

                    self._genArrival()


                if not EventList:
                    break

        print("calendar: \n")
        print("Total Events: ", self._total_events)
        print("Total Arrivals: ", self._generated_arrivals)
        print(calendar)


    def _genArrival(self):
            self._generated_arrivals +=1
            ia_time = np.random.exponential()
            #last_time = 
            arrival = self._time + ia_time
            sv_time = np.random.exponential()
            e = event(self._generated_arrivals, arrival, self.ARRIVAL , arrival, sv_time, ia_time)
            insort(EventList, e)

    def _EventHandler(self, e):
        self._time = e.event_time
        #Gestisco arrival
        if(e.event_type == self.ARRIVAL):
            if(not self._busy):
                print("Time: ", self._time)
                a = event(e.id, e.arrival_time, self.ARRIVAL, e.arrival_time, e.service_time, e.interarrival_time)
                print("Inserisco nel calendario la ", a.event_type, "del processo con id ", a.id)
                insort(calendar, a)
                self._busy = True
                d = event(e.id, e.event_time + e.service_time, self.DEPARTURE, e.arrival_time, e.service_time, e.interarrival_time)
                insort(EventList,d)
                #schedulo departure e aggiorno timer
            else:
                print("metto in coda")
                waiting_queue.append(e)
                #aggiungo alla coda e aggiorno timer

        #Gestisco Departure
        else:
            d = event(e.id, e.event_time, self.DEPARTURE, e.arrival_time, e.service_time, e.interarrival_time)
            print("Inserisco nel calendario la ", d.event_type, "del processo con id ", d.id)
            insort(calendar, d) 

            if(not waiting_queue):
                self._busy = False
            else:
                self._busy = False
                e = waiting_queue.pop(0)
                customer = event(e.id, self._time, self.ARRIVAL, e.arrival_time, e.service_time, e.interarrival_time)
                self._EventHandler(customer)
                #EventList.insert(0,customer)
            #aggiorno timer e metto busy false

    def _instance(self):
        self._time = 0
        self._busy = False
        self._generated_arrivals = 0 # mostly for id purposes
        self._total_events = 0
        self._maximum_events = 15 
```

```python
sim = Server()

sim.run()
```
